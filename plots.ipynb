```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot: average times for each problem group
groups = ['Binary', 'Permutation', 'Combinatorial']
rhc_times = [0.0, 0.03, 0.0]
sa_times = [0.0066, 0.315, 0.0]
ga_times = [0.148, 9.34, 0.21]
mimic_times = [16.594, 17.395, 23.17]

# X-axis locations for the groups
x = np.arange(len(groups))

# Bar width
width = 0.2

# Create plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for each algorithm
bars_rhc = ax.bar(x - width*1.5, rhc_times, width, label='RHC', color='blue')
bars_sa = ax.bar(x - width/2, sa_times, width, label='SA', color='orange')
bars_ga = ax.bar(x + width/2, ga_times, width, label='GA', color='green')
bars_mimic = ax.bar(x + width*1.5, mimic_times, width, label='MIMIC', color='red')

# Add labels, title, and legend
ax.set_xlabel('Problem Group')
ax.set_ylabel('Wall Clock Time (seconds)')
ax.set_title('Wall Clock Time Comparison Across Problem Groups')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()

# Display plot
plt.tight_layout()
plt.show()
```
